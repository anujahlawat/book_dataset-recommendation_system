### 3.similarity based algorithm 

```python
#1. sparse matrix
```

```python
#***************user - user similarity *****************

sparse_matrix = data1.pivot_table(index='User.ID',                           
                                 columns='Book.Title',             
                                 values='Book.Rating').reset_index(drop=True)  
sparse_matrix


#note : in sparse matrix all index are unique and all column names are unique


#*****************item - item similarity ****************
#here we go for user - user similarity method
#for item - item similarity just do transpose of matrix
#by interchanging index and columns values
```

```python
#Impute those NaNs with 0 values
sparse_matrix.fillna(0, inplace=True)
sparse_matrix
```

```python
#*********formation of user - user matrix **************
#entries of this matrix are filled by using cosine similarity score

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation

u_u_sim = 1 - pairwise_distances(sparse_matrix.values,metric='cosine')
u_u_sim
```

```python
#converting u_u_sim matrix into DataFrame
u_u_sim_df = pd.DataFrame(u_u_sim)
u_u_sim_df                                   #no. of records = no. of columns = no. of unique users
```

```python
#u_u_sim_df has natural numbers as index and columns names
#let replace them with user id's
u_u_sim_df.index = data1['User.ID'].unique()
u_u_sim_df.columns = data1['User.ID'].unique()

u_u_sim_df.head()
```

```python
np.fill_diagonal(u_u_sim, 0)
u_u_sim_df.head()
```
